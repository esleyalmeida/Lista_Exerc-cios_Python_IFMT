## 1) Maior e Menor Número
Leia 10 números inteiros e determine o maior e o menor sem usar `max()` ou `min()`.

```python
# Digite seu código aqui
2
# 1. ENTRADA E INICIALIZAÇÃO
maior = 0
menor = 0

for i in range(1, 11):
    num = int(input(f"Digite o {i}º número: "))

    # 2. PROCESSAMENTO
    if i == 1:
        # Ponto de partida: o primeiro número é o recordista inicial
        maior = num
        menor = num
    else:
        # Testes de recorde
        if num > maior:
            maior = num
        if num < menor:
            menor = num

# 3. SAÍDA
print(f"O maior número é: {maior}")
print(f"O menor número é: {menor}")
```
