## Exercício 7
Receba três números e mostre o maior.

```python
# entrada n1, n2 e n3 cada um em uma linha.
#processamento verificar qual dos numeros é o maior entre eles. teste SE o n1 é maior ou igual ao n2 E maior que n3, então o Maior é n1
#saída será mostrar o maior: print("O maior é: .")

n1 = float (input("Digite o Primeiro número: "))
n2 = float (input("Digite o Segundo número: "))
n3 = float (input("Digite o terceiro número: "))

if n1>=n2 and n1 >=n3:
  maior = n1
elif n2 >= n1 and n2 >= n3:
  maior = n2
else:
  maior = n3

print (f"O Maior número é: {maior}")
```

## Exercício 18
Conte vogais.

```python
# 1. ENTRADA
palavra = input("Digite uma palavra: ").lower() #(Convertemos para minúsculo)

# 2. GABARITO E CONTADOR
vogais = "aeiou"
contador = 0

# 3. PROCESSAMENTO (Letra por letra)
for letra in palavra:
    # 4. O TESTE: 'A letra está DENTRO de aeiou?'
    if letra in vogais:
        contador += 1

# 5. SAÍDA
print(f"A palavra tem {contador} vogais.")
```

## Exercício 19
Conte palavras de uma frase.

```python
# 1. ENTRADA
frase = input("Digite uma frase: ")

# 2. PROCESSAMENTO:
lista_de_palavras = frase.split() #O .split() cria uma lista com as palavras separadas
total_palavras = len(lista_de_palavras)

# 4. SAÍDA
print(f"A sua frase contém {total_palavras} palavras.")
```

## Exercício 21
Função que retorna maior número.

```python
# 1. DEFINIÇÃO DA FUNÇÃO
def encontrar_maior(a, b):
    if a > b:
        return a
    else:
        return b

# 2. USO DA FUNÇÃO
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

maior = encontrar_maior(num1, num2)

print(f"Entre {num1} e {num2}, o maior é: {maior}")
```

## Exercício 24
Remover duplicados da lista.

```python
#Entrada, a lista com dados repetidos
lista_original = [10, 20, 30, 40, 20, 50, 10]
#o set() remove os duplicados. o list() devolve para o formato de lista
nova_lista = list(set(lista_original))
print(f"Lista Original: {lista_original}")
print(f"nova lista: {nova_lista}")
```

## Exercício 25

Ordenar lista.

```python
#Dados inciais em lista []. lista = [valores]
lista = [3, 1, 4, 2, 7, 9]

# 2. PROCESSAMENTO, aqui A função sorted() cria uma nova versão organizada
lista_ordenada = sorted(lista)

#Saída
print(f"Lista Original: {lista}")
print(f"Lista Ordenada: {lista_ordenada}")
```

## Exercício 26
Contar ocorrência de número.

```python
# de ENTRADA eu preciso de uma lista de números (ex: [1, 2, 3, 2, 4, 2]) e o número que quero achar (ex: 2).
numeros = [10, 5, 20, 10, 45, 10, 30]
alvo = 10

#PROCESSAMENTO: preciso de uma varredura na lista do início ao fim, comparando cada item com o alvo e guarda o total de encontrados.
#Pra isso usar o comando .count()
quantidade = numeros.count(alvo)

# 3. SAÍDA
print(f"Lista: {numeros}")
print(f"O número {alvo} aparece {quantidade} vezes na lista.")
```
