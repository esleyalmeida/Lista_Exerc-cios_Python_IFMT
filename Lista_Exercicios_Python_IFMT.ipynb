## Exercício 2
Receba duas notas e calcule a média.

```python
#entrada, processamento e saída
# NOTAS = N1 e N2 (podem conter ".") por isso float | MÉDIA = B1 + B2 / 2 | IMPRIMIR MÉDIA
n1 = float(input("Digite a N1: "))
n2 = float(input("Digite a N2: "))
média = (n1+n2)/2
print(f"A média é: {média}")
```

## Exercício 3
Receba um número e mostre o dobro.

```python
# entrada, processamento e saída
# o que é preciso: receber um número (valor) e mostrar o resultado que será o dobro do valor.
# número = float(input("Digite um número: ")) < float pq o valor pode ser fracionado
# dobro é valor multiplicado por 2
# print(número)

numero = float(input("Digite um número: "))
dobro = numero * 2
print(f"O dobro de {numero} é {dobro}")
```

## Exercício 6
Receba um número e diga se é par ou ímpar.

```python

#meu raciocínio: Digitar uma valor de entrada. Um número.
#verificar se (condiçção) e o número digitar é par ou ímpar.
#saída: escrever se (condição) o valor digitado é par ou ímpar
#preciso de um teste para verificar a condição

número = int(input("Digite um número: "))
if número % 2 == 0:
  print ("é par")
else:
  print ("é impar")
```

## Exercício 7
Receba três números e mostre o maior.

```python
# entrada n1, n2 e n3 cada um em uma linha.
#processamento verificar qual dos numeros é o maior entre eles. teste SE o n1 é maior ou igual ao n2 E maior que n3, então o Maior é n1
#saída será mostrar o maior: print("O maior é: .")

n1 = float (input("Digite o Primeiro número: "))
n2 = float (input("Digite o Segundo número: "))
n3 = float (input("Digite o terceiro número: "))

if n1>=n2 and n1 >=n3:
  maior = n1
elif n2 >= n1 and n2 >= n3:
  maior = n2
else:
  maior = n3

print (f"O Maior número é: {maior}")
```
